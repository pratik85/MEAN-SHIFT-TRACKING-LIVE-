```python
import cv2
import os
 
from os.path import isfile, join
```

```python
def convert_frames_to_video(pathIn,pathOut,fps):
    frame_array = []
    files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
 
    #for sorting the file names properly
    print(len(files))
    for i in range(len(files)-1):
        filename=pathIn + files[i]
        #reading each files
        img = cv2.imread(filename)
        height, width, layers = img.shape
        size = (width,height)
        print(filename)
        #inserting the frames into an image array
        frame_array.append(img)
 
    out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
 
    for i in range(len(frame_array)):
        # writing to a image array
        out.write(frame_array[i])
    out.release()
```

```python
pathIn= './VOT CHALLENGE/'
pathOut = 'vot.avi'
fps = 25.0
convert_frames_to_video(pathIn, pathOut, fps)
```

```python
video = cv2.VideoCapture("./vot.avi")

_, first_frame = video.read()##video.read takes 2 value frame and truefalse
cv2.imshow("frame",first_frame)
cv2.waitKey(0)
print(first_frame.shape)
```

```python
x = 150
y = 87
width = 183
height = 145
roi = first_frame[y: y + height, x: x + width]
```

```python
hsv_roi = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)
roi_hist = cv2.calcHist([hsv_roi], [0], None, [360], [0, 360])## hsv_roi modell we want to convert to histogram ,0channel,360 bin,0-360 range of histogram  hue is basically cylindrical type of coordinate
roi_hist = cv2.normalize(roi_hist, roi_hist, 0, 255, cv2.NORM_MINMAX)

term_criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 1)
```

```python
while True:
    _, frame = video.read()
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    mask = cv2.calcBackProject([hsv], [0], roi_hist, [0, 360], 1)
    
    _, track_window = cv2.meanShift(mask, (x, y, width, height), term_criteria)
    x, y, w, h = track_window
    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
    
    cv2.imshow("Mask", mask)#mask means where the histogram is getting 
    cv2.imshow("Frame", frame)
    
    key = cv2.waitKey(30)
    if key == 27:
        break

video.release()
cv2.destroyAllWindows()
```
